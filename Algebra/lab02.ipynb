### 3. Підгрупи і гратка підгруп

Для кожної з груп $\mathbb{Z}_{24}$, $Q_8$, $D_5$, $S_4$ виконайте наступні кроки:

1. Створіть групу.
2. Знайдіть усі її підгрупи.   % G.subgroups()
3. Побудуйте гратку підгруп.   % G.subgroup_lattice() може не працювати, запитайте Gemini or ChatGpt
4. Вкажіть групу, яка сама не абелева, але всі її власні підгрупи є абелевими.



```python
from sage.all import *

def subgroup_lattice_manual(G):
    subs = G.subgroups()
    edges = []
    for H in subs:
        for K in subs:
            if H.is_subgroup(K) and H != K:
                if not any(H.is_subgroup(L) and L.is_subgroup(K) and L not in [H, K] for L in subs):
                    edges.append((H, K))
    lattice = DiGraph(edges)
    return lattice

def show_lattice_with_labels(G, name, figsize=[6,6], layout="spring"):
    print(f"\n=== Група {name} ===")
    print(f"Порядок: {G.order()}")
    print(f"Чи абелева? {G.is_abelian()}\n")
    
    subgroups = G.subgroups()
    print(f"Кількість підгруп: {len(subgroups)}")
    print("Підгрупи (порядок та генератори):")
    for i, H in enumerate(subgroups):
        print(f"  {i+1}. Порядок {H.order()}: {H.gens()}")

    lattice = subgroup_lattice_manual(G)
    
    labels = {H: i+1 for i, H in enumerate(lattice.vertices())}
    
    print("\nНумерація вершин для графа:")
    for H, i in labels.items():
        print(f"{i}: порядок {H.order()}")

    lattice.show(
        vertex_labels=labels,
        vertex_size=300,
        layout = layout,
        figsize=figsize,
    )
```

```python
Z24 = CyclicPermutationGroup(24)
show_lattice_with_labels(Z24, "Z_24", layout="circular")
```

```python
Q8 = QuaternionGroup()
print("Cама не абелева, але всі її власні підгрупи є абелевими.")
show_lattice_with_labels(Q8, "Q_8")
```

```python
# Ваш код тут:
D5 = DihedralGroup(5)
show_lattice_with_labels(D5, "D_5")
```

```python
# Ваш код тут:
S4 = SymmetricGroup(4)
show_lattice_with_labels(S4, "S_4", figsize=[9,9], layout='circular')
```

### 4. Системи твірних

Для кожної з груп $\mathbb{Z}_{24}$, $Q_8$, $D_5$, $S_4$ виконайте наступні кроки:

1. Створіть групу.
2. Знайдіть деяку систему твірних групи.    % G.gens()
3. Знайдіть всі мінімальні системи твірних групи.    % G.minimal_generating_sets()
4. Візьміть довільні два елементи групи, породіть ними підгрупу, знайдіть її порядок.  % G.subgroup([el1, el2])



```python
Z24 = AbelianGroup([24], names='a')  # a^24 = 1
a = Z24.gen(0)

print("=== Група Z_24 ===")
print("Порядок групи:", Z24.order())
print("Чи абелева?", Z24.is_abelian())

# Деяка система твірних
gens_Z24 = [a]
print("Деяка система твірних:", gens_Z24)

# Мінімальна система твірних (для циклічної групи завжди один елемент)
min_gens_Z24 = [gens_Z24]
print("Мінімальна система твірних:", min_gens_Z24)

# Породжена підгрупа двома елементами (a та a^2)
el1, el2 = a, a^2
H_Z24 = Z24.subgroup([el1, el2])
print(f"Породжена підгрупа елементами {el1}, {el2}:")
print("  Порядок підгрупи:", H_Z24.order())
print("  Генератори підгрупи:", H_Z24.gens())
```

Висновки

Кожна вершина має стільки вихідних ребер, скільки елементів у системі твірних.

Всі вершини мають однаковий ступінь.

Графи Келі дуже симетричні: будь\-яка вершина може "стати" будь\-якою іншою через ізоморфізм графа.

Для абелевих груп \(якщо взяти відповідну систему твірних\) графи часто виглядають як сітка або тор \(циклічна структура\).

Для неабелевих груп ребра "переплітаються", графи можуть виглядати складніше.

Орієнтований граф має 

∣G∣×∣S∣ ребер.

Якщо генератори включають обернені елементи, граф можна робити неорієнтованим.
