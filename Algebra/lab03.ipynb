### 1. Властивості циклічних груп

Для $n=100,173,530,746,911,1024$ розглянемо циклічну групу $\mathbb{Z}_n$ (або $C_n$):
1. Для кожного дільника $d$ порядку $n$:<br>
   1.1 Знайдіть кількість елементів порядку $d$ в групі $\mathbb{Z}_n$ (або $C_n$).<br>
   1.2 Порівняйте з $\phi(d)$.
2. Знайдіть кількість підгруп в групі $\mathbb{Z}_n$ (або $C_n$).<br>
   Порівняйте з $\tau(n)$.

Проаналізуйте результати та дайте відповіді на запитання:
1. Скільки елементів порядку $k$ в циклічній групі порядку $n$?
2. Cкільки підгруп в циклічній групі порядку $n$?

```python
# Ваш код тут:
n_lst = [100,173,530,746,911,1024]

for n in n_lst:
    divs = divisors(n)
    print("n =", n)
    print("divisors:", divs)
    print("d | phi(d) for divisors:")
    for d in divs:
        print(f"  {d} | {euler_phi(d)}")
    print("tau(n) =", len(divs))
    print("sum phi(d) over divisors =", sum(euler_phi(d) for d in divs))
    print("---")
```

1\)Якщо k не є дільником n, то в групі немає жодного елемента порядку k. Отже, кількість таких елементів  \- 0.

Якщо k є дільником n, то кількість елементів порядку k дорівнює $\phi(k)$. 

2\)Для кожного дільника d порядку n циклічної групи існує рівно одна підгрупа порядку $d$. 

Отже, кількість підгруп у циклічній групі порядку n дорівнює кількості натуральних дільників числа $n$.

### 2. Порядки елементів в дієдральних групах

Для $n=3,4,...,20$ розглянемо дієдральну групу $D_n$:   

1. Знайдіть кількість елементів порядку $d=2$ в групі $D_n$.
2. Для кожного дільника $d>2$ порядку $|D_n|=2n$:<br>
   Знайдіть кількість елементів порядку $d$ в групі $D_n$.

Проаналізуйте результати та дайте відповіді на запитання:

1. Скільки елементів порядку $d=2$ в дієдральній групі $D_n$?
2. Скільки елементів порядку $d$ в дієдральній групі $D_n$?
3. Чи для кожного дільника $d<2n$ порядку групи $D_n$ існує елемент порядку $d$?



```python
# Ваш код тут:

for n in range(3, 21):
    G = DihedralGroup(n)
    print(f"\n D_{n} (порядок групи = {G.order()}) ===")
    
    orders = sorted({g.order() for g in G})
    for d in orders:
        count = sum(1 for g in G if g.order() == d)
        print(f"Порядок {d:2d}: {count} елементів")
    
    if n % 2 == 0:
        expec_2 = n + 1
    else:
        expec_2 = n
    print(f"Кіількість елементів порядку 2: {expec_2}")
    
    print("Порядки:", orders)


print("\n Елементів порядку 2: n (якщо n непарне), n+1 (якщо парне).")
print(" Для d>2 і d|n: φ(d) елементів порядку d.")
print(" Якщо d не ділить n і d≠2 → таких елементів немає.")
print(" Не всі дільники 2n реалізуються як порядки елементів D_n.")
```

### 3. Дискретний логарифм в групі $\mathbb{Z}_n^{*}$

Розглянемо групу $\mathbb{Z}_n^{*}$ для $n=1102158078129785185997827239806$.

1. Перевірте, що група є циклічною і знайдіть її порядок.

3. Перевірте, що порядок елемента $\overline{a}$ дорівнює $\phi(n)$.
4. Виберіть три випадкових елементи $g_1,g_2,g_3$ групи $\mathbb{Z}_n^{*}$.
5. Для кожного $g$ знайдіть степінь $k$ такий, що $g=\overline{a}^k$.



```python
n = 1102158078129785185997827239806
G = Integers(n)

# Група Z_n* є циклічною, якщо n = 2, 4, p^k, або 2*p^k, де p - непарне просте
n_factors = factor(n)
is_cyclic = False
if len(n_factors) == 2 and n_factors[0][0] == 2 and n_factors[0][1] == 1 and is_prime(n_factors[1][0]):
    is_cyclic = True
elif len(n_factors) == 1 and (is_prime(n_factors[0][0]) or n_factors[0][0] in [2, 4]):
    is_cyclic = True

print(f"Група є циклічною? {is_cyclic}")

if not is_cyclic:
    print("Група не є циклічною, задача не може бути вирішена.")
else:
    phi_n = euler_phi(n)
    print(f"Порядок групи |G| = φ(n) = {phi_n}")
    
    order_factors = factor(phi_n)
    prime_divisors = [p for p, exp in order_factors]
    print(f"Прості дільники порядку: {prime_divisors}")

    a_val = None
    for a_cand in range(2, 100): 
        a = G(a_cand) 
        if gcd(a, n) != 1:
            continue
            
        is_generator_flag = True
        for q in prime_divisors:
            if power_mod(a, phi_n // q, n) == 1:
                is_generator_flag = False
                break
        
        if is_generator_flag:
            a_val = a
            print(f"\nЗнайдено найменший твірний елемент: a = {a_val}")
            break

    if a_val is not None:
        order_of_a = a_val.multiplicative_order()
        print(f"Порядок елемента a: {order_of_a}")
        print(f"Порядок збігається з φ(n)? {order_of_a == phi_n}")

        
        print("\nЗнаходження k для випадкових g = a^k:")
        for i in range(3):
            k_rand = ZZ.random_element(1, phi_n)
            g = power_mod(a_val, k_rand, n)
            g_elem = G(g)
            k_found = discrete_log(g_elem, a_val)            
                
            print(f"g = {g}")
            print(f"  Знайдений степінь k = {k_found}")
            print(f"  (Перевірка: початковий k був {k_rand})")
```

```python


A = matrix(ZZ, [[1,0,1],
                 [0,1,0],
                 [0,0,1]])
B = matrix(ZZ, [[1,0,0],
                 [0,1,1],
                 [0,0,1]])


def triple_to_mat(a,b,c):
    return matrix(ZZ, [[1,a,b],
                       [0,1,c],
                       [0,0,1]])

def mat_to_triple(M):
    return (ZZ(M[0,1]), ZZ(M[0,2]), ZZ(M[1,2]))

print("\nПеревірка відповідності генераторів у вигляді U(a,b,c):")
print("A triple:", mat_to_triple(A))
print("B triple:", mat_to_triple(B))

# (a,b,c) * (a',b',c') = (a+a', b+b' + a*c', c+c')
def mul_triples(t1, t2):
    a,b,c = t1
    a2,b2,c2 = t2
    return (a + a2, b + b2 + a*c2, c + c2)

tA = mat_to_triple(A)
tB = mat_to_triple(B)
Mprod = A * B
tprod = mul_triples(tA, tB)
print("\nA*B матрично =\n", Mprod)
print("A*B як трійки =", tprod, " -> матриця:\n", triple_to_mat(*tprod))

def comm_triples(t1, t2):
    # [t1,t2] = t1^{-1} t2^{-1} t1 t2 
    M1 = triple_to_mat(*t1)
    M2 = triple_to_mat(*t2)
    C = M1.inverse() * M2.inverse() * M1 * M2
    return mat_to_triple(C), C

comm_triple, comm_mat = comm_triples(tA, tB)
print("\nКомутатор [A,B] як трійка:", comm_triple)
print("Комутатор [A,B] матрично =\n", comm_mat)

#[(a,b,c),(a',b',c')] = (0, a*c' - a'*c, 0)
test_triples = [(1,2,3), (2,5,1), (0,7,0)]
for t1 in test_triples:
    for t2 in test_triples:
        Ctriple, Cmat = comm_triples(t1,t2)
        a,a1,b,b1,c,c1 = t1[0], t2[0], t1[1], t2[1], t1[2], t2[2]
        anal = (0, a*c1 - a1*c, 0)
        if Ctriple != anal:
            print("Неспівпадіння!", t1, t2, Ctriple, anal)
print("\nУсі збіглися")

#(a,b,c)^k = (k*a, k*b + a*c * k*(k-1)/2, k*c)
def power_triple(t, k):
    M = triple_to_mat(*t)
    Mk = M**k
    return mat_to_triple(Mk)

def analytic_power(t, k):
    a,b,c = t
    ka = k*a
    kc = k*c
    # k*b + a*c * k*(k-1)/2
    kb = k*b + a*c * (k*(k-1)//2)
    return (ka, kb, kc)

examples = [(1,0,0),(0,1,0),(1,2,3),(2,3,1)]
for t in examples:
    for k in [0,1,2,3,5,10]:
        p_mat = power_triple(t,k)
        p_anal = analytic_power(t,k)
        if p_mat != p_anal:
            print("Розбіжність у степені:", t, k, p_mat, p_anal)
print("\nФормула піднесення до степеня перевірена для прикладів.")
print("\n=== ВИСНОВКИ ===")
print("1) Кожен елемент G має вигляд U(a,b,c) = [[1,a,b],[0,1,c],[0,0,1]] з a,b,c ∈ Z.")
print("   Отже G ∼= Z^3 як множина, але з операцією (a,b,c)*(a',b',c') = (a+a', b+b' + a*c', c+c').")
print("2) Група нескінченна (має параметри a,b,c ∈ Z).")
print("3) Група некомутативна (комутатор не завжди тривіальний)")
print("4) Центр Z(G) = {(0,b,0) : b ∈ Z} ≅ Z (тобто матриці з a=c=0, довільним b).")
print("5) Всі ненульові елементи мають нескінченний порядок (бо їх трійкові координати зростають лінійно при піднесенні до степеня, або — елементи центру мають вид (0,b,0) і теж не мають скінченного порядку).")
print("6) Група не абелева і не циклічна. Ізоморфізм: Heisenberg група H(Z).")
```
